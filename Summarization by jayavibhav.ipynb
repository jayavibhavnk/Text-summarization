```python
#main
min_val=100

#text summarization
def textsum():
    global inputtxt,root
    root=Tk()
    root.geometry("600x600")
    root.configure(background = '#AA96DA')
    root.title("Summarizer")
    inputtxt = ScrolledText(root,height=10,
                   width = 50)
    inputtxt.pack()
    button0 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'Summarize',font=("Calibri",10), highlightcolor = '#ed815c', width = 20,height = 5,command=normal_text)
    button0.pack()

#normal summarization    
def normal_text():
    article=inputtxt.get('1.0', 'end-1c')
    print(article)
    summary=Text_sum(article)
    output_text=ScrolledText(root,height=10,width=50)
    output_text.insert(INSERT,summary)
    output_text.pack()

#functions for general summarization    
def Text_sum(article):
    global min_value
    min_value=100
    article=clean_data_before(article)
    summarizer1 = pipeline("summarization",model="t5-small")
    #article=article[:512]
    summary_text = summarizer1(article,do_sample=False,min_length=min_val)[0]['summary_text']
    summary_text = clean_data_after(summary_text)
    return(summary_text)

#image summarization
def imgsum():
    global inputtxt,root,label_file_explorer
    root=Tk()
    root.geometry("600x600")
    root.configure(background = '#AA96DA')
    root.title("Image Summarizer")
    label_file_explorer = Label(root,
                            text = "Select the image file",
                            width = 100, height = 4,
                            fg = "blue")
    label_file_explorer.pack()
    button0 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'Browse files',font=("Calibri",10), highlightcolor = '#ed815c', width = 20,height = 5,command=browseFiles)
    button0.pack()
    button1 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'Summarize',font=("Calibri",10), highlightcolor = '#ed815c', width = 20,height = 5,command=image_summary)
    button1.pack()


def img_to_sum(image):
    img = cv2.imread(image)
    text = pytesseract.image_to_string(img)
    k=Text_sum(text)
    return(k)

#image summary
def image_summary():
    img=filename
    summary=img_to_sum(img)
    output_text=ScrolledText(root,height=10,width=50)
    output_text.insert(INSERT,summary)
    output_text.pack()

#browse files on tkinter
def browseFiles():
    global filename
    filename = filedialog.askopenfilename(initialdir = "/",
                                          title = "Select a File",
                                          filetypes = (("Text files",
                                                        "*.txt*"),
                                                       ("all files",
                                                        "*.*")))
      
    # Change label contents
    label_file_explorer.configure(text="File Opened: "+filename)
    
    
    
    
    
#pdf summarization        
    
def pdfsum():
    global inputtxt,root,label_file_explorer
    root=Tk()
    root.geometry("600x600")
    root.configure(background = '#AA96DA')
    root.title("Image Summarizer")
    label_file_explorer = Label(root,
                            text = "Select the pdf file",
                            width = 100, height = 4,
                            fg = "blue")
    label_file_explorer.pack()
    button0 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'Browse files',font=("Calibri",10), highlightcolor = '#ed815c', width = 20,height = 5,command=browseFiles)
    button0.pack()
    button1 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'Summarize',font=("Calibri",10), highlightcolor = '#ed815c', width = 20,height = 5,command=pdf_summary)
    button1.pack()
    
#summary of pdf
def pdf_summary():
    pdf=filename
    summary=pdf_to_sum(pdf)
    output_text=ScrolledText(root,height=10,width=50)
    output_text.insert(INSERT,summary)
    output_text.pack()

#convert pdf to images and extract data using tesseract

def convert_pdf_to_img(pdf_file):
    pop = r"C:\Users\jayav\poppler-22.04.0\Library\bin"
    return convert_from_path(pdf_file,poppler_path=pop)


def convert_image_to_text(file):
    text = image_to_string(file)
    return text

def get_text_from_any_pdf(pdf_file):
    images = convert_pdf_to_img(pdf_file)
    final_text = ""
    for pg, img in enumerate(images):
        
        final_text += convert_image_to_text(img)
    return final_text

def pdf_sum_pre(path_to_pdf):    
    text=get_text_from_any_pdf(path_to_pdf)
    text = text.replace("-\n", "")
    l=[]
    for i in text.split():
        l.append(i)
    article=""
    for i in l:
        article+=i+" "
    return(article)

#pdf to summary function
def pdf_to_sum(path_to_pdf):
    text=pdf_sum_pre(path_to_pdf)
    k=Text_sum(text)
    return(k)    
        
    
#url summarization    

def urlsum():
    global inputtxt,root
    root=Tk()
    root.geometry("600x600")
    root.configure(background = '#AA96DA')
    root.title("URL summarization")
    inputtxt = ScrolledText(root,height=10,
                   width = 50)
    inputtxt.pack()
    button0 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'Summarize',font=("Calibri",10), highlightcolor = '#ed815c', width = 20,height = 5,command=url_output)
    button0.pack()
    
#output the summary from url
def url_output():
    url=inputtxt.get('1.0', 'end-1c')
    #print(url)
    summary=url_to_sum(url)
    #print(summary)
    output_text=ScrolledText(root,height=10,width=50)
    output_text.insert(INSERT,summary)
    output_text.pack()

#url to summary
def url_to_sum(URL):
    text=url_convert(URL)
    k=Text_sum(text[:512])
    return(k)

#web scrape url and obtain text
def url_convert(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    results = soup.find_all(['h1', 'p'])
    text = [result.text for result in results]
    ARTICLE = ' '.join(text)
    return(ARTICLE)    
    
    
#data cleaning

def clean_data_before(text):
    # Replacing all the occurrences of \n,\\n,\t,\\ with a space.
    Formatted_text = text.replace('\\n', ' ').replace('\n', ' ').replace('\t',' ').replace('\\', ' ').replace('. com', '.com').replace("    "," ")
    # Removing all the occurrences of links that starts with https
    Formatted_text = re.sub(r'http\S+', '', text)
    # Remove all the occurrences of text that ends with .com
    Formatted_text = re.sub(r"\ [A-Za-z]*\.com", " ", Formatted_text)
     # Remove accented characters from text using unidecode.
    # Unidecode() - It takes unicode data & tries to represent it to ASCII characters. 
    Formatted_text = unidecode.unidecode(Formatted_text)
    Formatted_text = Formatted_text.lower()
    #spell = Speller(lang='en')
    #Formatted_text = spell(Formatted_text)
    Formatted_Text = re.sub(r"[^a-zA-Z0-9:$-,%.?!]+", ' ', Formatted_text) 
    return Formatted_text

#capitalize after summarization
def clean_data_after(text):
    #capitalize
    s=""
    for i in text.split("."):
        s+=i.capitalize()+"."
    return(s)

#about menu
def about_menu():
    global inputtxt,root
    root=Tk()
    root.geometry("350x150")
    root.configure(background = '#AA96DA')
    root.title("About")
    output_text=ScrolledText(root,height=10,width=50)
    output_text.insert(INSERT,"""
    About
    
    Summarization using tansformers
    (normal)text, images, pdf, url
    
    Done by Jayavibhavnk""")
    output_text.pack()

#configure menu
def configure_menu():
    global min_val,w
    root=Tk()
    root.geometry("350x150")
    root.configure(background = '#AA96DA')
    root.title("configure")
    l1=Label(root,text="enter number of summary words")
    l1.pack()
    w = Spinbox(root, from_ = 25, to = 250)
    w.pack()
    b=Button(root,text="enter",command=gett_value)
    b.pack()

#spinbox value
def gett_value():
    global w,min_value
    min_value=w.get()
    print(min_value)

#main tkinter gui 
root=Tk()
root.geometry("600x600")
root.configure(background = '#AA96DA')
root.title("Summarization")
lb1=Label(root,background='#FFFFD2',text="Summarize",font=("Times New Roman",40))
lb1.grid(row = 0, column = 0, padx = 0, pady = 20)

button00 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'Text',font=("Calibri",10), highlightcolor = '#ed815c', width = 25,height = 5,command=textsum)
button00.grid(row = 1, column = 0, padx = 210, pady = 20)

button10 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'Image',font=("Calibri",10), highlightcolor = '#ed815c', width = 25,height = 5,command=imgsum)
button10.grid(row = 2, column = 0, padx = 210, pady = 10)

button20 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'PDF',font=("Calibri",10), highlightcolor = '#ed815c', width = 25,height = 5,command=pdfsum)
button20.grid(row = 3, column = 0, padx = 210, pady = 10)

button30 = Button(root,bg = '#C5FAD5', activebackground='#e8914f', text = 'URL',font=("Calibri",10), highlightcolor = '#ed815c', width = 25,height = 5,command=urlsum)
button30.grid(row = 4, column = 0, padx = 210, pady = 10)

menu = Menu(root)
root.config(menu=menu)
filemenu = Menu(menu)
menu.add_cascade(label='File', menu=filemenu)
filemenu.add_separator()
filemenu.add_command(label='Settings', command=configure_menu)
helpmenu = Menu(menu)
menu.add_cascade(label='Help', menu=helpmenu)
helpmenu.add_command(label='About',command=about_menu)
mainloop()

root.mainloop()

    
```
